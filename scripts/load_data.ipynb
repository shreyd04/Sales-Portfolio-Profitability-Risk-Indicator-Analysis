```python
import pandas as pd
from sqlalchemy import create_engine
import os

# 1. Path Setup
csv_path = "/Users/shreyd04/SuperStore Sales Dashboard/data/samplesuperstore.csv"

def load_data():
    try:
        # 2. Load with Latin-1 to handle special characters
        df = pd.read_csv(csv_path, encoding='latin1')
        
        # 3. Clean columns for SQL (removes spaces and hyphens)
        df.columns = [c.replace(' ', '_').replace('-', '_').lower() for c in df.columns]

        # 4. Connect using your Mac role 'shreyd04'
        # Ensure 'sales_db' exists in your PostgreSQL first!
        engine = create_engine('postgresql://shreyd04@localhost:5432/sales_db')

        # 5. Push to SQL
        df.to_sql('orders', engine, if_exists='replace', index=False)
        
        print("✅ SUCCESS: 10,000+ records migrated to PostgreSQL!")
        print(f"Verified Segment Sample: {df['segment'].unique()}")

    except Exception as e:
        print(f"❌ Error: {e}")

if __name__ == "__main__":
    load_data()
```
